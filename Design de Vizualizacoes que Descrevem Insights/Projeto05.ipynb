### Definição do Problema

O objetivo será analisar dados sobre os games disponíveis na Google PlayStore.

## Instalando e Carregando os Pacotes

```python
# Imports

import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import plotly.express as px
%matplotlib inline
sns.set()
```

```python
dados = pd.read_csv("dados/dataset4.csv")
```

```python
dados.shape
```

```python
dados.head()
```

```python
dados.info()
```

```python
data1 = dados.copy()
data1.installs.unique()
```

```python
install_range = {'500.0 M': 500000000, 
                 '100.0 M': 100000000, 
                 '50.0 M': 50000000, 
                 '10.0 M': 10000000, 
                 '5.0 M': 5000000, 
                 '1.0 M': 1000000,
                 '1000.0 M': 1000000000 , 
                 '500.0 k': 500000, 
                 '100.0 k': 1000000}
```

```python
data1['Installs_num'] = data1['installs'].map(install_range)
```

```python
data1
```

```python
total_install_cat = data1.groupby('category')['Installs_num'].sum().sort_values(ascending = False)
```

```python
total_install_cat
```

**Gráfico de barras na horizontal com o total de games e total de instalações por categoria**

```python
# Dataviz 27
plt.figure(figsize = (14, 8))
sns.barplot(y = 'category', x = 'Installs_num', data = data1)
plt.xlabel('Número de Games e Número de Instalações')
plt.ylabel('Categoria')
plt.show()
```

**Pie Chart com o percentual de games por categoria**

```python
# Dataviz 28
plt.figure(figsize = (14,7))
cat_labels = data1['category'].value_counts().index
plt.pie(data1['category'].value_counts().values, 
        labels = cat_labels, 
        autopct = '%1.1f%%', 
        startangle = 90,
        explode = [0.15, 0.12, 0.1, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08, 0.08])

plt.show()
```
