### Análise dos Métodos de Pagamento no Brasil

Este projeto analisa a evolução dos principais métodos de pagamento no Brasil ao longo do tempo, identificando mudanças estruturais no comportamento financeiro da população, como o crescimento do PIX e a redução de métodos tradicionais como DOC e cheque.

#### Instalação e Importação das Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configurar o estilo e o tamanho dos gráficos - "darkgrid" → fundo cinza com grid | "white" → fundo branco sem grid
sns.set_style("whitegrid") # Fundo branco (white) | Linhas de grade suaves (grid)
plt.rcParams["figure.figsize"] = (12,6) # rcParams Guarda configurações globais de plotagem | (12,6) → largura 12, altura 6
```

```python
df = pd.read_csv("dataset.csv")

df_valor = pd.read_csv("valor_cartao_credito.csv")
df_transacoes = pd.read_csv("Numero_Transacoes_Cartao_Credito.csv")
```

```python
#Backup dos Dados

df_backup = df.copy()
valor_backup = df_valor.copy()
trans_backup = df_transacoes.copy()
```

#### Limpeza e Padronização das Colunas

```python
# Renomeando as colunas p melhora legibilidade.

df_backup = df_backup.rename(columns={
    "YearMonth":"MesReferencia",
    "quantityPix":"NumeroTransacoesPIX",
    "valuePix":"ValorTotalPIX",
    "quantityTED":"NumeroTransacoesTED",
    "valueTED":"ValorTotalTED",
    "quantityTEC":"NumeroTransacoesTEC",
    "valueTEC":"ValorTotalTEC",
    "quantityBankCheck":"NumeroTransacoesCheque",
    "valueBankCheck":"ValorTotalCheque",
    "quantityBrazilianBoletoPayment":"NumeroTransacoesBoleto",
    "valueBrazilianBoletoPayment":"ValorTotalBoleto",
    "quantityDOC":"NumeroTransacoesDOC",
    "valueDOC":"ValorTotalDOC"
})
```

```python
df_backup.head()
```

```python
# Transformação da Data, e Criando colunas para Ano e Mês

df_backup["MesReferencia"] = pd.to_datetime(df_backup["MesReferencia"], format="%Y%m")

df_backup["Ano"] = df_backup["MesReferencia"].dt.year
df_backup["Mes"] = df_backup["MesReferencia"].dt.month
```

#### Engenharia de Atributos | Feature Engineering

```python
# Criar novas variáveis úteis. Valor médio de cada transação PIX

df_backup["ValorMedioPIX"] = (df_backup["ValorTotalPIX"] /df_backup["NumeroTransacoesPIX"]).fillna(0)
```

#### Análise Inicial

```python
df_backup.describe()
```

```python
df_backup.info()
```

```python
# Evolução Anual do PIX

pix_ano = (df_backup.groupby("Ano")["ValorTotalPIX"].sum().reset_index())
```

```python
pix_ano
```

```python
sns.lineplot(data=pix_ano, x="Ano", y="ValorTotalPIX", marker="o")

plt.title("Evolução Anual do PIX no Brasil")
plt.ylabel("Valor Total")
plt.xlabel("Ano")

plt.show()
```

📌 O PIX apresenta crescimento explosivo após 2020, indicando rápida adoção da tecnologia de pagamentos instantâneos no Brasil.

```python
# Comparação PIX vs Boleto

pix_boleto = df_backup.groupby("Ano")[["ValorTotalPIX","ValorTotalBoleto"]].sum().reset_index()
```

```python
pix_boleto
```

```python
pix_boleto.set_index("Ano").plot()

plt.title("PIX vs Boleto ao Longo do Tempo")
plt.ylabel("Valor Total")
plt.show()
```

📌 O PIX passou a competir diretamente com o boleto como método de pagamento digital.

📌 O PIX se tornou o principal método digital em poucos anos.

```python
# Declínio do DOC

doc_ano = df_backup.groupby("Ano")["ValorTotalDOC"].sum().reset_index()
```

```python
doc_ano
```

```python
sns.lineplot(data=doc_ano, x="Ano", y="ValorTotalDOC", marker="o", color="red")

plt.title("Declínio das Transações DOC")
plt.show()
```

📌 O DOC mostra queda contínua, indicando substituição por métodos mais rápidos como PIX e TED.

#### participação ao longo do tempo

```python
participacao = df_backup.groupby("Ano")[[
    "ValorTotalPIX",
    "ValorTotalTED",
    "ValorTotalBoleto",
    "ValorTotalCheque"
]].sum()

participacao
```

```python
participacao_pct = participacao.div(participacao.sum(axis=1), axis=0)
participacao_pct.plot.area()
```

📌 O crescimento do PIX está associado principalmente à queda do DOC e do cheque.

#### Correlação

```python
df_diff = df_backup[[
    "ValorTotalPIX",
    "ValorTotalTED",
    "ValorTotalBoleto",
    "ValorTotalCheque"
]].pct_change()

# pct_change() — variação percentual - calcula o crescimento percentual. 
```

```python
correlacao = df_diff.corr()

sns.heatmap(correlacao, cmap="coolwarm", annot=True, fmt=".2f")

plt.title("Correlação das Variações entre Métodos de Pagamento")
plt.show()
```

📌 O crescimento do PIX apresenta correlação negativa com métodos tradicionais como DOC e cheque, sugerindo substituição desses meios de pagamento.

📌 Os métodos digitais apresentam padrões de crescimento semelhantes, indicando uma transição do sistema financeiro para pagamentos eletrônicos.

#### Importação dos Dados de Cartão de Crédito

```python
# Importação dos Dados de Cartão de Crédito | tipo de junção:how="inner" - anos que existem em ambos os DataFrames.

df_cartao = pd.merge(valor_backup, trans_backup, on="Ano")

df_cartao.head()
```

```python
# Padronização dos Tipos

df_cartao = df_cartao.astype({"valorcartaocredito":float,"NumeroTransacoesCartaoCredito":float})
```

```python
# Integração com Dataset Principal

df_final = pd.merge(df_backup, df_cartao, on="Ano")
```

```python
df_final.head()
```

```python
df_final.shape
```

#### Participação dos Métodos de Pagamento

```python
# Selecionando último ano.

ultimo_ano = df_final["Ano"].max()

dados_participacao = df_final[df_final["Ano"] == ultimo_ano][[
    "ValorTotalPIX",
    "ValorTotalTED",
    "ValorTotalBoleto",
    "valorcartaocredito"
]].sum()
```

```python
dados_participacao
```

```python
dados_participacao.plot(
    kind="pie",
    autopct="%1.1f%%",
    ylabel=""
)

plt.title("Participação dos Métodos de Pagamento")
plt.show()
```

#### Valor médio por transação

```python
df_backup["ValorMedioTED"] = df_backup["ValorTotalTED"] / df_backup["NumeroTransacoesTED"]
df_backup["ValorMedioPIX"] = df_backup["ValorTotalPIX"] / df_backup["NumeroTransacoesPIX"]
```

```python
df_backup[["ValorMedioTED","ValorMedioPIX"]].head()
```

📌 TED possui ticket médio maior que PIX.

📌 Os dados mostram que o PIX se consolidou como principal meio de pagamento para microtransações, enquanto o TED permanece relevante para transferências de alto valor.

📌 Isso indica que o PIX é amplamente utilizado para pagamentos de pequeno valor do dia a dia, enquanto o TED continua sendo utilizado para transferências de maior valor, como movimentações empresariais ou transferências bancárias mais relevantes.

## 📊 Principais Insights do Projeto
#### 1️⃣ Crescimento explosivo do PIX

Após seu lançamento em 2020 pelo Banco Central do Brasil, o PIX apresentou crescimento extremamente acelerado, tornando-se rapidamente um dos principais meios de pagamento do país.

Os dados indicam que, em poucos anos, o sistema passou de adoção inicial para ampla utilização, evidenciando uma rápida aceitação da população e do mercado financeiro.

#### 2️⃣ Declínio de métodos tradicionais

Métodos de transferência bancária mais antigos, como DOC e cheque, apresentam queda consistente ao longo do período analisado.

Esse comportamento sugere um processo de substituição tecnológica, onde meios de pagamento mais lentos e burocráticos estão sendo gradualmente substituídos por soluções digitais mais eficientes.

#### 3️⃣ Transformação estrutural no sistema de pagamentos

A análise evidencia uma mudança estrutural no sistema financeiro brasileiro, caracterizada pela migração para métodos de pagamento:

>instantâneos
>
>digitais
>
>de baixo custo operacional

Esse movimento reforça a tendência de digitalização dos serviços financeiros, impulsionada pela inovação tecnológica e pela infraestrutura de pagamentos instantâneos implementada no país.

#### 📌 O PIX não apenas cresceu rapidamente, mas também passou a competir diretamente com métodos consolidados como boleto e TED, indicando uma mudança significativa no comportamento de pagamento da população.
