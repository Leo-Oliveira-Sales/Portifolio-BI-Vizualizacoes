# Como tratar as tabelas:

* Os dados de Cartão de Credito foram extraidos de dados abertos do Banco Central através do link abaixo:
>https://www.bcb.gov.br/estatisticas/spbadendos?ano=2024
* A planilha do Banco Central do Brasil normalmente vem em formato horizontal, por isso tivemos de tratar a tabela
* Tivemos tambem que excluir as colunas e linhas NaN
* Depois Salvamos o Dataset pronto

```python
import pandas as pd

arquivo = "Dados Estatísticos 2024S1.xlsx"

df_qtd = pd.read_excel(arquivo, sheet_name="2.1.a")
df_valor = pd.read_excel(arquivo, sheet_name="2.1.b")
```

```python
df_valor.head()
```

```python
df_qtd.head()
```

```python
# Transpor os dados
df_valor = df_valor.T

# Agora colunas viram linhas.
```

```python
# Transformar a primeira linha em cabeçalho

df_valor.columns = df_valor.iloc[0]
df_valor = df_valor[1:]
```

```python
# Transformar índice em coluna Ano

df_valor = df_valor.reset_index()
df_valor.rename(columns={"index": "Ano"}, inplace=True)
```

```python
# remover colunas totalmente vazias
df_valor = df_valor.dropna(axis=1, how="all")

# pegar a linha que tem os dados
dados = df_valor.iloc[1]

# pegar os anos das colunas
anos = df_valor.columns

# criar novo dataframe
df_final = pd.DataFrame({
    "Ano": anos,
    "valorcartaocredito": dados.values
})

# remover linhas que não são anos
df_final = df_final[pd.to_numeric(df_final["Ano"], errors="coerce").notnull()]

# converter tipos
df_final["Ano"] = df_final["Ano"].astype(int)
df_final["valorcartaocredito"] = pd.to_numeric(df_final["valorcartaocredito"], errors="coerce")

display(df_final)
```

```python
# Corrigindo o indice

df_final = df_final.reset_index(drop=True)
```

```python
df_final
```

```python
# Salvando

df_final.to_csv("valor_cartao_credito.csv", index=False)
```
